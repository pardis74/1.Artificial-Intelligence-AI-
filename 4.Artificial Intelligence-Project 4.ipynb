```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# --- Set Georgia font ---
rcParams['font.family'] = 'Georgia'

# --- Membership Functions ---
def high_temperature(x, a=30, b=40, c=50):
    return np.piecewise(x,
                        [x <= a, (a < x) & (x <= b), (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - a) / (b - a), 1, 0])

def high_energy_consumption(x, b=100, c=150):
    return np.piecewise(x,
                        [x <= b, (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - b) / (c - b), 1])

# --- Input range ---
x = np.linspace(0, 150, 1000)

# --- Membership values ---
temp_high = high_temperature(x)
energy_high = high_energy_consumption(x)
rule_output = np.minimum(temp_high, energy_high)

# --- Plotting ---
plt.figure(figsize=(10, 6))
plt.plot(x, temp_high, label='High Temperature', linestyle='--', color='red')
plt.plot(x, energy_high, label='High Energy Consumption', linestyle='-.', color='blue')
plt.plot(x, rule_output, label='Rule Output: Reduce Energy Consumption', color='green')

# --- Decorations ---
plt.title('Fuzzy Rule: IF Temperature High AND Energy High THEN Reduce Energy')
plt.xlabel('Input Level')
plt.ylabel('Membership Degree')
plt.ylim(-0.05, 1.05)
plt.legend()
plt.grid(True)

# --- Save and Show ---
plt.savefig("/Users/pardissadatian/Documents/New-Folder/4.1.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# --- Set Georgia font ---
rcParams['font.family'] = 'Georgia'

# --- Membership Functions ---
def high_humidity(x, b=60, c=80):
    return np.piecewise(x,
                        [x <= b, (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - b) / (c - b), 1])

def high_energy_consumption(x, b=100, c=150):
    return np.piecewise(x,
                        [x <= b, (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - b) / (c - b), 1])

# --- Input Ranges ---
humidity_values = np.linspace(0, 100, 1000)
energy_values = np.linspace(0, 150, 1000)

# For simplicity, align ranges by interpolation
humidity_scaled = np.interp(humidity_values, (0, 100), (0, 150))

# --- Membership values ---
humidity_high = high_humidity(humidity_values)
energy_high = high_energy_consumption(energy_values)
combined_membership = np.minimum(humidity_high, high_energy_consumption(humidity_scaled))

# --- Plot ---
plt.figure(figsize=(10, 6))
plt.plot(humidity_values, humidity_high, label='High Humidity', linestyle='--', color='red')
plt.plot(humidity_values, high_energy_consumption(humidity_scaled), label='High Energy Consumption (rescaled)', linestyle='-.', color='blue')
plt.plot(humidity_values, combined_membership, label='Rule Output: Reduce Energy Consumption', color='green')

# --- Decorations ---
plt.title('Fuzzy Rule: IF Humidity High AND Energy High THEN Reduce Energy')
plt.xlabel('Humidity Level (scaled to Energy Consumption)')
plt.ylabel('Membership Degree')
plt.ylim(-0.05, 1.05)
plt.legend()
plt.grid(True)

# --- Save and Show ---
plt.savefig("/Users/pardissadatian/Documents/New-Folder/4.2.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Membership Functions ---

# Temperature (High)
def high_temperature(x, a=30, b=40, c=50):
    return np.piecewise(x,
                        [x <= a, (a < x) & (x <= b), (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - a) / (b - a), 1, 0])

# Energy Consumption (Low, Medium, High)
def low_energy_consumption(x, a=0, b=50):
    return np.piecewise(x,
                        [x <= a, (a < x) & (x < b), x >= b],
                        [1, lambda x: (b - x) / (b - a), 0])

def medium_energy_consumption(x, a=50, b=100, c=150):
    return np.piecewise(x,
                        [x <= a, (a < x) & (x <= b), (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - a) / (b - a), lambda x: (c - x) / (c - b), 0])

def high_energy_consumption(x, b=100, c=150):
    return np.piecewise(x,
                        [x <= b, (b < x) & (x < c), x >= c],
                        [0, lambda x: (x - b) / (c - b), 1])

# Humidity (High)
def high_humidity(y, b=60, c=80):
    return np.piecewise(y,
                        [y <= b, (b < y) & (y < c), y >= c],
                        [0, lambda y: (y - b) / (c - b), 1])

# Time of Day (Evening)
def evening_time(t, b=16, c=20):
    return np.piecewise(t,
                        [t <= b, (b < t) & (t < c), t >= c],
                        [0, lambda t: (t - b) / (c - b), 1])

# Training Progress (High)
def high_training_progress(t, b=70, c=100):
    return np.piecewise(t,
                        [t <= b, (b < t) & (t < c), t >= c],
                        [0, lambda t: (t - b) / (c - b), 1])

# Past Performance (Good)
def good_performance(m, b=70, c=100):
    return np.piecewise(m,
                        [m <= b, (b < m) & (m < c), m >= c],
                        [0, lambda m: (m - b) / (c - b), 1])

# ML Speed (Fast)
def fast_speed(s, b=70, c=100):
    return np.piecewise(s,
                        [s <= b, (b < s) & (s < c), s >= c],
                        [0, lambda s: (s - b) / (c - b), 1])


# --- Input ranges ---
x_temp = np.linspace(0, 150, 1000)
x_energy = np.linspace(0, 150, 1000)
x_humidity = np.linspace(0, 100, 1000)
x_time = np.linspace(0, 20, 1000)
x_training = np.linspace(0, 100, 1000)
x_performance = np.linspace(0, 100, 1000)
x_speed = np.linspace(0, 100, 1000)


# --- Rule 1: IF (Temperature is High) AND (Energy Consumption is High) THEN (Reduce Energy Consumption) ---
rule1 = np.minimum(high_temperature(x_temp), high_energy_consumption(x_energy))
sum_rule1 = np.sum(rule1)

plt.figure(figsize=(10,6))
plt.plot(x_temp, high_temperature(x_temp), label='High Temperature', linestyle='--')
plt.plot(x_energy, high_energy_consumption(x_energy), label='High Energy Consumption', linestyle='-.')
plt.plot(x_temp, rule1, label='Rule 1 Output', color='purple')
plt.title('Rule 1: IF Temperature is High AND Energy Consumption is High THEN Reduce Energy Consumption')
plt.xlabel('Level')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()


# --- Rule 2: IF (Humidity is High) AND (Energy Consumption is High) THEN (Reduce Energy Consumption) ---
rule2 = np.minimum(high_humidity(x_humidity), high_energy_consumption(x_energy))
sum_rule2 = np.sum(rule2)

plt.figure(figsize=(10,6))
plt.plot(x_humidity, high_humidity(x_humidity), label='High Humidity', linestyle='--')
plt.plot(x_energy, high_energy_consumption(x_energy), label='High Energy Consumption', linestyle='-.')
plt.plot(x_humidity, rule2, label='Rule 2 Output', color='red')
plt.title('Rule 2: IF Humidity is High AND Energy Consumption is High THEN Reduce Energy Consumption')
plt.xlabel('Level')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()


# --- Rule 3: IF (Time of Day is Evening) AND (Energy Consumption is High) THEN (Reduce Energy Consumption) ---
rule3 = np.minimum(evening_time(x_time), high_energy_consumption(x_energy))
sum_rule3 = np.sum(rule3)

plt.figure(figsize=(10,6))
plt.plot(x_time, evening_time(x_time), label='Evening Time', linestyle='--')
plt.plot(x_energy, high_energy_consumption(x_energy), label='High Energy Consumption', linestyle='-.')
plt.plot(x_time, rule3, label='Rule 3 Output', color='green')
plt.title('Rule 3: IF Time of Day is Evening AND Energy Consumption is High THEN Reduce Energy Consumption')
plt.xlabel('Level')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()


# --- CO2 Emissions Rule 1: IF (Training Progress is High) AND (Past Performance is Good) THEN CO2 Emissions is Low ---
rule4 = np.minimum(high_training_progress(x_training), good_performance(x_performance))
sum_rule4 = np.sum(rule4)

plt.figure(figsize=(10,6))
plt.plot(x_training, high_training_progress(x_training), label='High Training Progress', linestyle='--')
plt.plot(x_performance, good_performance(x_performance), label='Good Past Performance', linestyle='-.')
plt.plot(x_training, rule4, label='Rule 4 Output', color='orange')
plt.title('Rule 4: IF Training Progress is High AND Past Performance is Good THEN CO2 Emissions is Low')
plt.xlabel('Level')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()


# --- CO2 Emissions Rule 2: IF (Speed of ML Algorithm is Fast) AND (Energy Consumption of ML Algorithm is Low) THEN CO2 Emissions is Low ---
rule5 = np.minimum(fast_speed(x_speed), low_energy_consumption(x_energy))
sum_rule5 = np.sum(rule5)

plt.figure(figsize=(10,6))
plt.plot(x_speed, fast_speed(x_speed), label='Fast ML Algorithm Speed', linestyle='--')
plt.plot(x_energy, low_energy_consumption(x_energy), label='Low Energy Consumption', linestyle='-.')
plt.plot(x_speed, rule5, label='Rule 5 Output', color='brown')
plt.title('Rule 5: IF Speed is Fast AND Energy Consumption is Low THEN CO2 Emissions is Low')
plt.xlabel('Level')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()


# --- Aggregation and Output ---

best_membership_energy = max(sum_rule1, sum_rule2, sum_rule3)
best_membership_co2 = max(sum_rule4, sum_rule5)

output_layer_mean = (best_membership_energy + best_membership_co2) / 2
sigmoid_output = 1 / (1 + np.exp(-output_layer_mean))

print("Sum Rule 1 (Energy):", sum_rule1)
print("Sum Rule 2 (Energy):", sum_rule2)
print("Sum Rule 3 (Energy):", sum_rule3)
print("Best Membership (Energy):", best_membership_energy)
print()
print("Sum Rule 4 (CO2):", sum_rule4)
print("Sum Rule 5 (CO2):", sum_rule5)
print("Best Membership (CO2):", best_membership_co2)
print()
print("Output Layer Mean:", output_layer_mean)
print("Sigmoid Output (Probability):", sigmoid_output)
```
