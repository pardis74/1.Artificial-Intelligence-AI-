```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x values (speeds in miles per hour) from 0 to 3000
x = np.linspace(0, 3000, 100)

# Define the membership functions for "Very Low" and "High"
def very_low_membership(x):
    return 1 - (x / 3000)

def high_membership(x):
    return (x - 2000) / 1000

# Calculate the membership values for "Very Low" and "High" for the given x values
membership_very_low = very_low_membership(x)
membership_high = high_membership(x)

# Calculate the fuzzy intersection for each x as the minimum of membership values
fuzzy_intersection = np.minimum(membership_very_low, membership_high)

# Create a plot
plt.figure(figsize=(8, 6))
plt.plot(x, fuzzy_intersection, label="Fuzzy Intersection (Very Low Ç High)", color="purple")
plt.xlabel("Speed (miles per hour)")
plt.ylabel("Membership Value")
plt.title("Fuzzy Intersection: Very Low ^ High")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the membership function for "Very Low" and "Complement of High" (not high)
def membership_VeryLow(x):
    return 1 - (x / 3000)

def membership_Complement_High(x):
    return 1 - ((x - 2000) / 1000)

# Define the fuzzy union function
def fuzzy_union(x):
    mu_VeryLow = membership_VeryLow(x)
    mu_Complement_High = membership_Complement_High(x)
    
    # Calculate the fuzzy union without using min or max
    mu_Union = mu_VeryLow + mu_Complement_High - mu_VeryLow * mu_Complement_High
    
    return mu_Union

# Create an array of speed values (x) within the specified range
x_values = np.linspace(0, 3000, 100)  # Adjust the number of points as needed

# Calculate the fuzzy union for each speed value
y_values = [fuzzy_union(x) for x in x_values]

# Create a plot to visualize the fuzzy union
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Fuzzy Union (Very Low and not High)')
plt.xlabel('Speed (mph)')
plt.ylabel('Membership Value')
plt.title('Fuzzy Union of "Very Low" and "not High"')
plt.legend()
plt.grid(True)
plt.show()
```
